```python
# modules we need

import numpy as np  
import matplotlib.pyplot as plt  
import h5py

# Load h5py file
ds = h5py.File("test.h5", "r")
# see what is inside that h5 file
print(ds.keys())
```

```python
x = ds['x']
organ = ds['organ']
```

```python
print(x.shape)
```

```python
plt.subplots(3,4,figsize = (20,20))

for i in range(12):
    img_np = x[i]
    plt.subplot(3,4,1+i)
    plt.imshow(img_np)
plt.show()
```

```python
import random
from PIL import Image

# Select 100 random indices from the range of the number of images in x
random_indices = random.sample(range(x.shape[0]), 500)

# Iterate through the selected indices and save each image to disk
for i, index in enumerate(random_indices):
    img = x[index]
    img = Image.fromarray(img)
    img.save("image_{}.jpg".format(i))
```
